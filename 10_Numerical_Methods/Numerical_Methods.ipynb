### –û–±—Ä–∞–±–æ—Ç–∫–∞ –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π

````python
def numerical_hists(data):
    columns = data.dtypes[data.dtypes != "object"].index

    fig, axs = plt.subplots(len(columns), 1, figsize=(10, 20))

    axi = 0

    for column in columns:
        axs[axi].hist(data[column], bins=50)
        axs[axi].set_title(column)
        axi += 1
````

````python
numerical_hists(data)
````

1) –¥–ª—è –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏ –∞–≤—Ç–æ –æ–ø—Ä–µ–¥–µ–ª–∏–º —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: –º–æ–¥—ã –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∞–≤—Ç–æ (—Ç–∏–ø, –∫–æ—Ä–æ–±–∫–∞, —Ç–æ–ø–ª–∏–≤–æ –∏ –ø—Ä.), –º–µ–¥–∏–∞–Ω—ã –¥–ª—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω (—Ü–µ–Ω–∞).
2) –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –º–æ–¥–µ–ª–∏ –∞–≤—Ç–æ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —Ç–∞–±–ª–∏—Ü–µ–π –∏–∑ –ø. 1.

````python
data["MakeModel"] = data["Brand"] + "_" + data["Model"]
````

````python
cartype_modes = data.pivot_table(
    values=["VehicleType", "Gearbox", "Power", "FuelType"],
    index="MakeModel",
    aggfunc=lambda column: column.mode()
)
````

–í—ã–±–µ—Ä–µ–º –∏–∑ –Ω–µ–≥–æ –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç.

````python
def select_last_item(array):
    
    if type(array) == np.ndarray:  ## –ï—Å–ª–∏ –º–æ–¥—É —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å –Ω–µ –ø–æ–ª—É—á–∞–µ—Ç—Å—è (–≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è —É–Ω–∏–∫–∞–ª—å–Ω—ã),
                                   ## mode() –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å–æ –≤—Å–µ–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
        if len(array) > 0:         ## –ï—Å–ª–∏ –ø–æ —Ç–∏–ø—É –∞–≤—Ç–æ –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö (–≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è NaN),
            return array[-1]       ## mode() –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫
        else:
            return np.NaN
    else:
        return array               ## –ï—Å–ª–∏ —Ä–∞—Å—á–µ—Ç –º–æ–¥—ã —É–¥–∞—á–µ–Ω –∏ –ø–æ–ª—É—á–µ–Ω–æ –æ–¥–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ
````

````python
cartype_modes["Power"] = cartype_modes["Power"].apply(lambda item: select_last_item(item)).astype("float32")
cartype_modes["VehicleType"] = cartype_modes["VehicleType"].apply(lambda item: select_last_item(item))
````

–ú–æ–¥—ã —Ä–∞—Å—Å—á–∏—Ç–∞–ª–∏, —Ç–µ–ø–µ—Ä—å —Ä–∞—Å—á–∏—Ç–∞–µ–º –º–µ–¥–∏–∞–Ω—ã –¥–ª—è —Ü–µ–Ω—ã –∏ –æ–±—ä–µ–¥–∏–Ω–∏–º –≤—Å–µ –≤ –æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—É

````python
cartype_medians = data.pivot_table(
    values="Price",
    index="MakeModel",
    aggfunc="median"
)
````

````python
cartype_summary = cartype_modes.merge(cartype_medians, how="left", left_index=True, right_index=True)
````

–¢–µ–ø–µ—Ä—å –º–æ–∂–Ω–æ –∑–∞–ø–æ–ª–Ω—è—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –ø–æ —Ç–∞–±–ª–∏—Ü–µ.

````python
for column in cartype_summary.columns:
    intermediate = data[column]
    intermediate.index = data["MakeModel"]
    intermediate.fillna(cartype_summary[column], inplace=True)
    intermediate.index = data.index
    data[column] = intermediate
````
