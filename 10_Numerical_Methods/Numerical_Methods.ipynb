### Заполнение пропусков

Еще раз посмотрим, сколько пропусков есть.

```python
print(data.info())
```

Пропусков много, сбросить их не удастся.

Попробуем заполнить следующим образом:
1) для каждой модели авто определим характерные параметры: моды для параметров авто (тип, коробка, топливо и пр.), медианы для непрерывных величин (цена).
2) заполним пропуски в зависимости от модели авто в соответствии с таблицей из п. 1.

```python
data["MakeModel"] = data["Brand"] + "_" + data["Model"]
```

```python
cartype_modes = data.pivot_table(
    values=["VehicleType", "Gearbox", "Power", "FuelType"],
    index="MakeModel",
    aggfunc=lambda column: column.mode()
)
```

Моды есть, но не везде, там где моду вычислить невозможно, в таблицу попал список.

Выберем из него последний элемент.

```python
display(cartype_modes[cartype_modes.index == "land_rover_serie_2"])
```

```python
def select_last_item(array):
    
    if type(array) == np.ndarray:  ## Если моду рассчитать не получается (все значения уникальны),
                                   ## mode() возвращает список со всеми значениями
        if len(array) > 0:         ## Если по типу авто нет данных (все значения NaN),
            return array[-1]       ## mode() возвращает пустой список
        else:
            return np.NaN
    else:
        return array               ## Если расчет моды удачен и получено одно значение
```

```python
cartype_modes["Power"] = cartype_modes["Power"].apply(lambda item: select_last_item(item)).astype("float32")
cartype_modes["VehicleType"] = cartype_modes["VehicleType"].apply(lambda item: select_last_item(item))
```

Моды рассчитали, теперь расчитаем медианы для цены и объединим все в одну таблицу

```python
cartype_medians = data.pivot_table(
    values="Price",
    index="MakeModel",
    aggfunc="median"
)
```

```python
cartype_summary = cartype_modes.merge(cartype_medians, how="left", left_index=True, right_index=True)
```

Теперь можно заполнять пропуски по таблице.

```python
for column in cartype_summary.columns:
    intermediate = data[column]
    intermediate.index = data["MakeModel"]
    intermediate.fillna(cartype_summary[column], inplace=True)
    intermediate.index = data.index
    data[column] = intermediate
```

```python
print(data.info())
```

Теперь пропусков стало существенно меньше.

Остаются два интересующих нас столбца со значительной (>5%) долей пропусков - Repaired и Model.

В столбце Repaired заполним данные случайными флагами так, чтобы после заполнения пропорция сохранилась. Думаю, это весьма валидный подход, т.к. распределение битых машин должно быть относительно равномерным вне зависимости от параметров авто.

```python
nonrepaired_part = (
    len(data[data["Repaired"] == "no"])
    /
    (len(data[data["Repaired"] == "no"]) + len(data[data["Repaired"] == "yes"]))
)

print("Доля неремонтированных машин в данных: {0:.4f}".format(nonrepaired_part))
```

```python
data["Repaired"] = data["Repaired"].fillna(
    pd.Series(np.random.choice(["no", "yes"], size=shape_init[0], p=[nonrepaired_part, 1 - nonrepaired_part]))
)
```

Не вижу, как можно относительно просто заполнить модели, поэтому, я думаю, оставить как есть. В любом случае я их буду кодировать, вероятнее всего, через OHE, поэтому, просто NaNам выделится отдельный столбец.

Остальных пропусков немного, их сбросим.

```python
data = data.dropna(subset=["Price", "VehicleType", "RegistrationYear", "Gearbox", "Power", "FuelType"])
```

## Обучение моделей

Прежде чем обучать модели, нужно сделать 2 вещи:
1. Закодировать категориальные переменные
2. Отскейлить имеющиеся данные, т.к. одна из моделей, которую я хочу протестировать - kNN.

### Скейлинг и кодировка данных

```python
numerical_columns = ["RegistrationYear", "Power", "Kilometer"]
categorial_columns = ["VehicleType", "Gearbox", "Model", "FuelType", "Brand", "Repaired"]

transformer = make_column_transformer(
    (StandardScaler(), numerical_columns),
    (OneHotEncoder(drop="first"), categorial_columns)
)
```

```python
transformer.fit(features_train);
```

```python
features_train_transformed = transformer.transform(features_train)
```

Данные подготовлены, можно обучать модели.

### Базовая модель

Прежде чем обучать настоящие модели, стоит понять, где проходит граница адекватности для точности. Для этого обучим простую модель.

```python
def rmse(truths, predictions):
    return mean_squared_error(truths, predictions) ** .5
```

```python
baseline_model = GridSearchCV(
    DummyRegressor(),
    param_grid={},
    scoring=make_scorer(rmse, greater_is_better=False),
    n_jobs=-1
)
```

```python
%%time
baseline_model.fit(features_train_transformed, target_train)
```

```python
baseline_model.best_score_
```

Граница адекватности ясна, ожидаю, что RMSE для всех обученных моделей будет ниже.

Еще будем собирать сводную таблицу для финального сравнения.

```python
models_summary = pd.DataFrame({
    "RMSE":[-baseline_model.best_score_],
    "fit_time":[.766],
    "predict_time":[np.NaN]
                              }
)
```
