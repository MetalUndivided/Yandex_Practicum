### Получение данных

#### Импорты библиотек

```python
import pandas as pd
import numpy as np
import re
import datetime
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
## Для открытия на локальной машине

try:
    df_shows = pd.read_csv("/datasets/mkrf_shows.csv", sep=",")
    df_movies = pd.read_csv("/datasets/mkrf_movies.csv", sep=",")
    
except FileNotFoundError:
    df_shows = pd.read_csv("mkrf_shows.csv", sep=",")
    df_movies = pd.read_csv("mkrf_movies.csv", sep=",")
```

#### Первичный осмотр

Некоторые типы стоит конвертировать:
* *puNumber* в int64, заодно переименуем столбец
* *show_start_date* в datetime
* *ratings* в float64

Столбец *puNumber* переименуем прямо сейчас в обоих датафреймах

```python
for df in [df_shows, df_movies]:
    df.rename({"puNumber":"pu_number"}, axis=1, inplace=True)
```

### Объединение датафреймов

```python
## в pu_number затаилась запись "нет", она всего одна и мешает конвертации, дропнем ее
## если не сконвертировать, смержить не получится
df_movies.drop(df_movies.query("pu_number == 'нет'").index, inplace=True)
df_movies["pu_number"] = df_movies["pu_number"].astype("int64")
```

```python
temp = df_movies.merge(df_shows, "left", "pu_number")
```

```python
#display(temp)
```

```python
df_movies = temp
print(df_movies.info())
```

### Удаление дубликатов

#### Удаление явных дубликатов

На всякий случай удалим явные дубликаты, сделать это нужно сейчас, пока над данным не проведено никаких манипуляций и явные дубликаты не превратились в неявные.

```python
shows_shape_init = df_shows.shape
movies_shape_init = df_movies.shape

df_shows.drop_duplicates(inplace=True)
df_movies.drop_duplicates(inplace=True)

print(
"""
DF shows
Удалено дубликатов: {0}

DF movies
Удалено дубликатов: {1}
""".format(
    shows_shape_init[0] - df_shows.shape[0],
    movies_shape_init[0] - df_movies.shape[0]
)
     )
```

Явных дубликатов не оказалось.

#### Поиск и удаление неявных дубликатов

```python
def print_names(df):
    temp = list(df["title"].unique())
    temp.sort()
    print(temp)
```

```python
print_names(df_movies)
```

Неявные дубликаты есть, но вот выявить какую-то надежную закономерность их появления довольно сложно.

Пока что в глаза бросаются 3 основных типа:
* пробелы/точки/нижние подчеркивания в конце названий, например *'Хаббл 3D', 'Хаббл 3D.', 'Хаббл 3D_'*
* год где-то может отсутствовать, а где-то присутствовать, например *'Эверест', 'Эверест (2015)', 'Эверест .'*
* одно и то же название может быть как полным, так и сокращенным, например *'Элизиум - рай не на Земле', 'Элизиум.'*

Все типы ошибок, видимо, вызваны ручным заполнением данных.

Первый тип относительно легко обрабатывается регексами

Второй тип тоже легко обрабатывается регексами, но тут есть риск утери информации, если вдруг окажется, что это вовсе не дубликаты

Третий тип безопасно регексами обработать не удастся, но может эти дубликаты совпадают по номеру прокатного удостоверения?

```python
temp = df_movies.pivot_table(index="pu_number", values="title", aggfunc="count")
display(temp.sort_values("pu_number", ascending=False))
```

Неуникальных номеров нет, похоже с третьей категорией ничего поделать не удастся.

```python
rgx = re.compile("\(\d{4}\)")    ## Матчит год в скобках
df_movies["title"] = df_movies["title"].str.replace(rgx, "")

rgx = re.compile("[ \._-]+(?=$)")  ## Матчит пробел, точку, нижнее подчеркивание и дефис в конце стринга
df_movies["title"] = df_movies["title"].str.replace(rgx, "")

rgx = re.compile(" +")            ## Заодно заменим множественные пробелы одиночными
df_movies["title"] = df_movies["title"].str.replace(rgx, " ")

print_names(df_movies)
```

Несколько дубликатов остается (тип 3), но заменять их будет рискованно, есть шанс непредвиденно превратить одни фильмы в другие.

Пока что оставим эти дубликаты как есть.

### Конвертация типов данных

Конвертируем *pu_number* и *show_start_date*

```python
df_movies["show_start_date"] = pd.to_datetime(df_movies["show_start_date"], format="%Y-%m-%dT%H:%M:%S.%fZ")
#print(df_movies.info())
```

Выясним, почему рейтинги записаны как object

```python
print(df_movies["ratings"].sort_values(ascending=False).head(15))
```

Часть значений записана как процентные, потому и object. Конвертируем их.

```python
df_movies["ratings"] = df_movies.apply(lambda row: float(row["ratings"].rstrip("%")) / 100 if "%" in str(row["ratings"]) else row["ratings"], axis=1)
df_movies["ratings"] = df_movies["ratings"].astype("float64")
print(df_movies.info())
```

Теперь типы данных соответствуют содержащимся значениям.

В столбцах *ratings* и *box_office* теоретически можно заменить пропуски на некоторые средние значения, но я вовсе не уверен, что это будет правильное решение. Все-таки сильно разносортные у нас данные: тут и отдельные серии сериалов, и российское кино, и зарубежное, и современное, и второй половины прошлого века и т.д.

Теперь самое интересное - фильмы с гос. финансированием. Пока что выглядит так, что объемы финансирования указаны только для фильмов, для которых указан источник финансирования. Проверим это.

```python
display(df_movies[df_movies["financing_source"].isna()][["refundable_support", "nonrefundable_support", "budget", "financing_source"]])
display(df_movies[~(df_movies["financing_source"].isna())][["refundable_support", "nonrefundable_support", "budget", "financing_source"]])
```

Да, так и есть. Заполнять пропуски не будем, просто будем подразумевать, что там, где источник финансирования не указан, его не было, соответственно, и объемы финансирования там равны нулю.

```python
## для отслеживания объема удаленных данных
support_entries_init = df_movies[~(df_movies["nonrefundable_support"].isna())].shape
#print(support_entries_init)
```

```python
limits = {
    "refundable_support":.7e8,
    "nonrefundable_support":1.2e8,
    "budget":.4e9
}

for column, limit in limits.items():
    dummy = df_movies[df_movies[column] > limit].index
    df_movies.drop(dummy, inplace=True)
    print("Дропнуто {0} строк, превышающих граничное значение столбца {1}".format(len(dummy), column))
```

```python
def track_losses(init: int, current: int) -> None:
    print(
    """
    Начальное количество строк: {0}
    Конечное количество строк: {1}
    Дропнуто строк: {2}
        Что составляет {3:.2%} от начального количества
    """.format(
        init,
        current,
        init - current,
        (init - current) / init
    )
)
```

```python
support_entries_fin = df_movies[~(df_movies["nonrefundable_support"].isna())].shape

track_losses(support_entries_init[0], support_entries_fin[0])
```

```python
def get_first_item(df, source: str) -> pd.Series:
    
    ## Accepts a pandas.DataFrame and a title of the column to be split
    ## Returns pandas.Series that contains the first item of the specified column
    
    dest = df[source].str.split(",")
    dest = dest.apply(lambda row: row[0])
    
    return dest
```

```python
sns.scatterplot(data=films_supported, x="support_part", y="box_office", hue="budget");
corr = films_supported["support_part"].corr(films_supported["box_office"])
print("Коэф. корреляции составляет {0:.3f}".format(corr))
```

Из этого графика можно сделать 2 заключения:
1. Бюджет фильма является более точным предиктором кассовых сборов фильма, чем объем гос. финансирования.
2. Большинство фильмов с высокой долей гос. поддержки являются малобюджетными, а следовательно, собирают невысокие кассы.

Проверим, может между абсолютным объемом поддержки и сборами есть зависимость?

```python
sns.scatterplot(data=films_supported, x="total_support", y="box_office", hue="budget");
corr = films_supported["total_support"].corr(films_supported["box_office"])
print("Коэф. корреляции составляет {0:.3f}".format(corr))
```

Некая зависимость есть, но все же выглядит так, что бюджет здесь имеет основной вес.

Зная, что от между бюджетом фильма и его кассовыми сборами есть взаимосвязь, стоит все же проверить взаимосвязь рейтинга фильма от его бюджета.

#### Повторная проверка зависимости рейтинга фильма от объема гос. поддержки

Перестроим график зависимости рейтинга от бюджета, но в этот раз раскрасим точки в зависимости от бюджета фильма.

```python
sns.scatterplot(data=films_supported, x="support_part", y="ratings", hue="budget");
corr = films_supported["support_part"].corr(films_supported["ratings"])
print("Коэф. корреляции составляет {0:.3f}".format(corr))
```

Возможно есть некая обратная зависимость, т.к. фильмы с более высоким бюджетом будто бы имеют большие рейтинги?

Проверим это, построив кроссплот рейтинга от бюджета, цвету точек будет соответствовать доля гос. поддержки.

```python
sns.scatterplot(data=films_supported, x="budget", y="ratings", hue="support_part");
corr = films_supported["budget"].corr(films_supported["ratings"])
print("Коэф. корреляции составляет {0:.3f}".format(corr))
```

Не похоже, чтобы была какая-то значимая зависимость. Скорее даже наоборот: после определенной точки размера бюджета (250Е6) рейтинг уменьшается с увеличением бюджета.

Первоначальная теория остается в силе: рейтинг не зависит от объемов гос. поддержки.
